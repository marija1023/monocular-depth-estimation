U folderu *splits* se cuvaju prethodno zapamcene putanje do frejmova koje se dele u tri foldera - train, val, test.

```
def save_paths(paths):
  imgs_len = len(paths)
  test_split_point = 4 * imgs_len // 5
  val_split_point = 4 * test_split_point // 5
  train_paths = paths[:val_split_point]
  val_paths = paths[val_split_point:test_split_point]
  test_paths = paths[test_split_point:]

  train_file = os.path.join(split_folder, 'train_files.txt')
  val_file = os.path.join(split_folder, 'val_files.txt')
  test_file = os.path.join(split_folder, 'test_files.txt')

  with open(train_file, "a") as trf:
    for t in train_paths:
      trf.write("{0}\n".format(t))

  with open(val_file, "a") as vf:
    for t in val_paths:
      vf.write("{0}\n".format(t))

  with open(test_file, "a") as tef:
    for t in test_paths:
      tef.write("{0}\n".format(t))
```

```
save_paths(main_img_paths)
```
