```
img_root_dir = '/content/drive/My Drive/ML/Projekat/monocular-depth-estimation/data/photos'
img_dirs = list(filter(lambda x: x[0] != '\.', os.listdir(img_root_dir)))
img_dirs
```

```
def load_photos(path):
  # img_dir = list(filter(lambda x: x == 'data', os.listdir(os.path.join(path, 'image_02'))))[0]
  images = [cv2.imread(file) for file in glob(os.path.join(path, 'image_02/data/*.png'))]
  return list(map(lambda x: Image.fromarray(x, 'RGB'), images))
```

```
def get_photo_path(path):
  imgs_path = os.listdir(os.path.join(path, 'image_02/data'))
  return list(map(lambda x: os.path.join(path, 'image_02/data', x), imgs_path))
```

```
img_paths = []

for dir in img_dirs:
  img_paths.append(get_photo_path(os.path.join(img_root_dir, dir)))
```
